```
sean_map = np.zeros([filtered_image.shape[0], filtered_image.shape[1]], dtype = float) 
```

```
sean_map[:][sean_map.shape[0]-1] = filtered_image[:][sean_map.shape[0]-1]
```

```
sean_map
```

```
for i in range(dim[0]-2,0,-1):
  for j in range(0,dim[1]):
    if(j==0):
      sean_map[i][j] = filtered_image[i][j]+min(sean_map[i+1][j],sean_map[i+1][j+1])
    elif(j==dim[1]-1):
      sean_map[i][j] = filtered_image[i][j]+min(sean_map[i+1][j],sean_map[i+1][j-1])
    else:
      sean_map[i][j] = filtered_image[i][j]+min(min(sean_map[i+1][j],sean_map[i+1][j-1]),sean_map[i+1][j+1])

      
  
```
